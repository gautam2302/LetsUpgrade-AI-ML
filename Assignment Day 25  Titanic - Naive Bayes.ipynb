```python
data1=data.drop(['Cabin', 'Ticket','Name'], axis=1)
data1.head()
```

```python
enc = preprocessing.LabelEncoder()
data1['Sex'] = enc.fit_transform(data1.Sex)
data1['Embarked'] = enc.fit_transform(data1.Embarked)
```

```python
data1.head()
```

```python
def prediction(col):
    
    clf = BernoulliNB()

    for i in col:
        x = data1.drop([i], axis=1)
        print("####################################################################################")
        print("                                                                                    ")
        print(f"Here Dependent Variable is {i} & remaining are Independent Variables")
        print("x :\n", x.head())
        y = data1[i]
        print("\ny:\n", y.head())
    
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.3, random_state=0)
        model = clf.fit(x_train,y_train)
        pred = model.predict(x_test)
    
        print(f"\nAcuuracy for {i} is: ", accuracy_score(y_test, pred, normalize=True))
        print(f"Confusion matrix for {i} is \n", confusion_matrix(y_test, pred))
        print("                                                                                    ")

col = ['Survived', 'Pclass', 'Sex', 'SibSp', 'Parch','Embarked']        
prediction(col)
```
